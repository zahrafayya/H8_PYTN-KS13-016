```python
# Soal 9
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('mpg')

df['temp'] = np.where(df['horsepower'] < 105, 'lower', 'upper') 

df['horse_origin'] = df['temp'] + " " + df['origin']

df.drop('temp', axis=1, inplace=True)

df.head()
```

```python
# Soal 10
import seaborn as sns
import pandas as pd

df = sns.load_dataset('mpg')

result = (df.groupby('origin')['weight'].describe())[['mean', 'std', 'count', 'mean','max']]

test = pd.DataFrame({'sum' : df.groupby('origin')['weight'].sum()}).reset_index()


final = pd.merge(result, test, on='origin')
final
```
