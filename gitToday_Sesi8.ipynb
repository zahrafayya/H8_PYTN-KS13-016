```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium

pd.set_option('display.max_columns', 500)

df = pd.read_csv('melb_data.csv')
```

```python
# Soal 8
df = df.sort_values(by=['Price'], ascending=False)
df = df.head(10)
world_map = folium.Map(location=[-37.8136, 144.9631], zoom_start=11)

incidents = folium.map.FeatureGroup()

for lat, long, price, addr in zip(df["Lattitude"], df["Longtitude"], df["Price"], df["Address"]):
    folium.Marker([lat, long], popup="Price: "+str(price), tooltip=addr).add_to(world_map)

    
world_map
    
```
