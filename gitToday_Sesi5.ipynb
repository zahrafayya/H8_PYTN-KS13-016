```python
# Soal 5
import numpy as np
import pandas as pd

series = pd.Series(np.random.randint(1,5,size=(100)))

print(series)
print("\n")

occurence = series.value_counts()

print("Jumlah angka:")
print(occurence)
print("\n")

print("Dua paling tinggi:")
print(occurence.nlargest(2))
```
